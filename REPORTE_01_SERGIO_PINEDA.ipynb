```python
def bar_chart(data, x , y ,color,text,title_gral):
    """
    La funcion se encarga de graficar con una grafica de barras la informacion enviada 
    [] todo - Renderizar grafica en la celda que genera la llamada  
    """
    bars = alt.Chart(data).mark_bar().encode(
        x='x:Q',
        y=alt.Y("y:O", sort='x'),
        color=alt.Color(f'{color}:Q', legend=alt.Legend(title="Score"),scale=alt.Scale(scheme='bluegreen',type= 'sqrt')),
    )

    text = bars.mark_text(
       align='left',
        baseline='middle',
        dx=3 
    ).encode(
        text=f'{text}:Q')
    return (bars + text).properties(title=title_gral)
```
